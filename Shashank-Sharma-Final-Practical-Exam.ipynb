# Practical Examination
### Download the dataset grapes_new.csv from https://mitu.co.in/dataset and perform the following operations.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

#### 1. Load the dataset

```python
df = pd.read_csv('dataset/grapes_new.csv')
df
```

#### 2. Separate the input and output variables

```python
x = df.drop('CLASS', axis=1)
y = df['CLASS']
print('Input Variables: ', list(x.columns))
print('Output Variable: CLASS')
```

#### 3. Apply the data cleaning (if necessary)

```python
# Removing NaN
print('Variables with NaN values:')
print(x.count()[x.count() < 194])
x['RIPENESS_PER'] = x['RIPENESS_PER'].fillna(x['RIPENESS_PER'].mean())
x['ALCOHOL_PER'] = x['ALCOHOL_PER'].fillna(x['ALCOHOL_PER'].mean())
x['ALCALINITY_OF_ASH'] = x['ALCALINITY_OF_ASH'].fillna(x['ALCALINITY_OF_ASH'].mean())
x['HUE'] = x['HUE'].fillna(x['HUE'].mean())
x['WATER_O2_PER'] = x['WATER_O2_PER'].fillna(x['WATER_O2_PER'].mean())

print('\n\nValues after removing NaN:')
print(x.count())
```

#### 4. Explore the output variable through countplot, pie chart

```python
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.title('Countplot of Class variable')
sns.countplot(df['CLASS']);
plt.subplot(1,2,2)
plt.title('Piechart of Class variable')
df.groupby('CLASS').size().plot(kind='pie', textprops={'fontsize': 15},
                                  colors=['tomato', 'gold', 'skyblue']);
```

#### 5. Explore the shape and soil type through countplot, pie chart

```python
plt.figure(figsize=(16,9))
plt.subplot(2,2,1)
plt.title('Countplot of Shape variable')
sns.countplot(df['SHAPE']);
plt.subplot(2,2,2)
plt.title('Countplot of Soil Type variable')
sns.countplot(df['SOIL_TYPE']);
plt.subplot(2,2,3)
plt.title('Piechart of Shape variable')
df.groupby('SHAPE').size().plot(kind='pie', textprops={'fontsize': 15},
                                  colors=['tomato', 'gold'])
plt.subplot(2,2,4)
plt.title('Piechart of Soil Type variable')
df.groupby('SOIL_TYPE').size().plot(kind='pie', textprops={'fontsize': 15},
                                  colors=['tomato', 'gold'])
```

#### 7. Apply the data encoding (if necessary)

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(x['SHAPE'])
encoded_labels = le.transform(x['SHAPE'])
x['SHAPE'] = encoded_labels
le.fit(x['SOIL_TYPE'])
encoded_labels = le.transform(x['SOIL_TYPE'])
x['SOIL_TYPE'] = encoded_labels
le.fit(y)
y = le.transform(y)
```

#### 8. Select only COLOR_INTENSITY FLAVANOIDS PROLINE TEMPERATURE FER_P2O5_PER as input features.

```python
from sklearn.model_selection import train_test_split
x = x[['COLOR_INTENSITY', 'FLAVANOIDS', 'PROLINE', 'TEMPERATURE', 'FER_P2O5_PER']]
y = y
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0)
```

#### 9. Implement classification algorithm

```python
x['Label'] = y
sns.pairplot(x, hue='Label');
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
x_train.columns
```

```python
rf = RandomForestClassifier()
rf.fit(x_train, y_train)
```

#### 10. Show prediction on some unseen data.

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred_rf = rf.predict(x_test)
accuracy_score(y_pred_rf, y_test)
```
