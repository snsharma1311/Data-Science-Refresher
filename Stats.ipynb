```python
def zscore(n):
    z = (n - n.mean()) / n.std()
    return z
```

```python
zscore(num1)
```

```python
q1, q2, q3 = np.percentile(salary, 25), np.percentile(salary, 50), np.percentile(salary, 75)
q1, q2, q3
```

```python
iqr = q3 - q1
iqr
```

```python
upper_fence = q3 + 1.5 * iqr
lower_fence = q1 - 1.5 * iqr
upper_fence, lower_fence
```

```python
sal = np.array(salary)
sal > upper_fence
```
