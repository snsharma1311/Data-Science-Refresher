### Part 1: Use the given file named cricket.csv and perform the following operations:

#### 1. Read the file in DataFrame

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('dataset/cricket.csv')
df.head()
```

#### 2. List the name of cricketer and their respective runs

```python
df[['name', 'runs']]
```

#### 3. Find total wickets taken by them

```python
print('Total Wickets: ', df['wickets'].sum())
```

#### 4. Find average of catches taken

```python
print('Average of Catches Taken: ', df['catches'].mean())
```

#### 5. Find the name of wicketkeeper

```python
print('Wicketkeeper: ', list(df.loc[df['stumpings'] > 0]['name'])[0])
```

#### 6. Print the name of bowler who played highest number of matches

```python
df_bowler = df.loc[(df['wickets'] > 0) & (df['stumpings'] == 0)]
print('Bowler with highest number of matches: ', list(df_bowler.loc[df_bowler['matches'] == df_bowler['matches'].max()]['name'])[0])
```

#### 7. Find average of all the bowlers

```python
df_bowler.insert(7, 'average', df['wickets'] / df['matches'], True)
```

```python
df_bowler[['name', 'average']]
```

#### 8. Find name of the bowler with least bowling average (wickets per match).

```python
print('Bowler with least average: ', list(df_bowler.loc[df_bowler['average'] == df_bowler['average'].min()]['name'])[0])
```

#### 9. Print information about all players whose catches per match ratio is greater than 0.5 store it in a csv file.

```python
df_catches_per_match = df.loc[(df['catches'] / df['matches']) > 0.5]
df_catches_per_match.to_csv('dataset/cricket_catches_per_match.csv')
```

#### 10. Draw the bar chart of player name against their runs.

```python
df.plot.bar(x='name', y='runs')
```

#### 11. Sort and print information about players by ascending order of runs

```python
df.sort_values(by=['runs'])
```

#### 12. Print the names of players whose wickets are greater than matches

```python
print('Players with wickets > matches: ', list(df.loc[df['wickets'] > df['matches']]['name']))
```

### Part 2: Download the file of fb.csv which contains the information about different posts information posted on Facebook page.

```python
df_fb = pd.read_csv('dataset/fb.csv')
df_fb
```

#### 1. Count how many number of posts are paid and unpaid?

```python
print('Number of paid posts are: ', len(df_fb.loc[df_fb['Paid'] > 0]))
print('Number of unpaid posts are: ', len(df_fb.loc[df_fb['Paid'] == 0]))
print('Number of posts for which information is not available: ', len(df_fb) - df_fb['Paid'].count())
```

#### 2. Display only Page type, Month, likes, comments and shares.

```python
df_fb[['Type', 'Post Month', 'comment', 'like', 'share']]
```

#### 3. Find average number of likes to the post in September month.

```python
print('Average number of likes in September: ', round(df_fb.loc[df_fb['Post Month'] == 9]['like'].mean(), 3))
```

#### 4. Print the information about the posts which are links.

```python
df_fb.loc[df_fb['Type'] == 'Link']
```

#### 5. How many total shares are present for the video posts?

```python
print('Total shares for Video posts: ', df_fb.loc[df_fb['Type'] == 'Video']['share'].sum())
```

#### 6. In which hour of the day maximum posts are posted?

```python
print('The hour of the day when maximum posts are posted: ', list(df_fb.groupby('Post Hour').agg({'Post Hour': 'count'}).idxmax())[0])
```

#### 7. In which day of October, maximum number of interactions are found?

```python
print('The day of October with maximum number of interactions: ', list(df_fb.loc[df_fb['Post Month'] == 10][['Total Interactions', 'Post Weekday']].groupby('Post Weekday').agg({'Total Interactions': 'sum'}).idxmax())[0])
```

#### 8. Print the information of the posts where likes are above 200 and shares are above 30.

```python
df_fb.loc[(df_fb['like'] > 200) & (df_fb['share'] > 30)]
```

#### 9. Which kind of posts are posted maximum number of times in November month?

```python
print('The kind of posts are posted maximum number of times in November month: ', list(df_fb.loc[df_fb['Post Month'] == 9].groupby('Type').agg({'Type': 'count'}).idxmax())[0])
```

#### 10. Find average and median of interactions for the paid (1) posts.

```python
print('Average of interactions for the paid (1) posts:', round(df_fb.loc[df_fb['Paid'] == 1]['Total Interactions'].mean(), 3))
print('Median of interactions for the paid (1) posts:', df_fb.loc[df_fb['Paid'] == 1]['Total Interactions'].median())
```
