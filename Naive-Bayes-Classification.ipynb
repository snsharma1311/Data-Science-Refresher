```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dataset/Social_Network_Ads.csv')
```

```python
df
```

```python
x = df[['Age', 'EstimatedSalary']]
y = df['Purchased']
```

```python
y.value_counts()[1]
```

```python
def purchase_probability(age, salary, df):
    similarity_ind_age = 15
    similarity_ind_salary = 40000
    # p(purchase | x) = p(x | purchase) * p(purchase) / p(x)
    p_purchase = df['Purchased'].value_counts()[1]/len(df['Purchased'])
    similar_x = df.loc[(df['Age'] >= age - similarity_ind_age) & (df['Age'] <= age + similarity_ind_age) & (df['EstimatedSalary'] >= salary - similarity_ind_salary) &  (df['EstimatedSalary'] <= salary + similarity_ind_salary)]
    p_x = len(similar_x) / len(df)
    p_x_purchase = similar_x['Purchased'].value_counts()[1]/len(similar_x)
    p_purchase_x = p_x_purchase * p_purchase / p_x
    return p_purchase_x
```

```python
purchase_probability(45, 40000, df)
```
